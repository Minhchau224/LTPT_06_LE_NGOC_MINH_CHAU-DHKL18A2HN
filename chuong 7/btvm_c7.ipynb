# I. Câu hỏi thảo luận

### 1. Tkinter: Là thư viện tiêu chuẩn của Python dùng để tạo giao diện đồ họa người dùng (GUI).


### 2. 
Chức năng của Tkinter trong python: Cung cấp các công cụ (widgets) như nút bấm, ô nhập liệu, bảng biểu để xây dựng ứng dụng Windows.
Các bước để tạo một tkinter app: 
(1) Import thư viện
(2) Tạo cửa sổ chính (root)
(3) Thêm các widget
(4) Chạy vòng lặp sự kiện

### 3. Root tkinter: Là cửa sổ chính (cửa sổ cha) chứa tất cả các thành phần khác của ứng dụng. Được tạo bằng lệnh root = Tk()

### 4. Chức năng phương thức Python Tkinter pack(): Sắp xếp các widget theo các khối (trên, dưới, trái, phải) bằng cách lấp đầy không gian trống.

### 5. Chức năng của phương thức grid(): Sắp xếp các widget theo cấu trúc bảng (dòng và cột).

### 6. Chức năng của phương thức place(): Đặt widget vào vị trí chính xác bằng cách sử dụng tọa độ (x, y).

### 7. Tham số của pack(): có 3 tham số chính: side, fill, expand

### 8. Tham số của grid(): có 4 tham số chính thường dùng: row, column, padx/pady, sticky

### 9. Tham số của place(): có 4 tham số chính thường dùng: x, y, width, heigt (hoặc relx, rely)

### 10. Nhận phản hồi API vào Textbox: sử dụng thư viện requests để gọi API, sau đó dùng phương thức textbox.insert (END, data) để hiển thị kết quả

### 11. Lấy chỉ mục (index) trong Combobox: Sử dụng phương thức combobox.current()

### 12. 
Nhúng hình ảnh vào Canvas bằng PIL:

(1) Mở ảnh bằng Image.open().

(2) Chuyển sang định dạng Tkinter bằng ImageTk.PhotoImage().

(3) Vẽ lên canvas bằng canvas.create_image().

### 13. Hiển thị ngày hiện tại: Sử dụng thư viện datetime để lấy thời gian, sau đó cập nhật vào một Label bằng tham số text hoặc textvariable.

### 14. 
Tạo thư mục mới với askdirectory: * Import filedialog.

Dùng path = filedialog.askdirectory() để lấy đường dẫn người dùng chọn.

Dùng os.makedirs(path) để tạo thư mục.

### 15.
Sử dụng StringVar trong Entry:

Khai báo var = StringVar().

Gán vào Entry: Entry(root, textvariable=var).

Dùng var.get() để lấy dữ liệu và var.set() để thay đổi nội dung ô nhập.

### Bài 7.1.

```python
import tkinter as tk
from tkinter import messagebox

def dang_nhap():
    user = entry_user.get()
    password = entry_pass.get()
    
    if user == "admin" and password == "123":
        messagebox.showinfo("Thông báo", "Đăng nhập thành công!")
    else:
        messagebox.showerror("Lỗi", "Sai tài khoản hoặc mật khẩu")

window = tk.Tk()
window.title("Cửa sổ đăng nhập")  
window.geometry("300x200")

tk.Label(window, text="Tài khoản:").pack(pady=5)
entry_user = tk.Entry(window)
entry_user.pack()

tk.Label(window, text="Mật khẩu:").pack(pady=5)
entry_pass = tk.Entry(window, show="*")
entry_pass.pack()

tk.Button(window, text="Đăng nhập", command=dang_nhap).pack(pady=15)

window.mainloop()
```

### Bài 7.4.

```python
import tkinter as tk

def kiem_tra():
    if entry_name.get() and entry_id.get() and entry_pass.get():
        btn_submit.pack(pady=10)
    else:
        btn_submit.pack_forget()

window = tk.Tk()
window.title("Thông tin sinh viên")
window.geometry("350x250")

tk.Label(window, text="Student Name:").pack(anchor="w", padx=20)
entry_name = tk.Entry(window)
entry_name.pack(fill="x", padx=20)

tk.Label(window, text="Student ID:").pack(anchor="w", padx=20, pady=(10,0))
entry_id = tk.Entry(window)
entry_id.pack(fill="x", padx=20)

tk.Label(window, text="Password:").pack(anchor="w", padx=20, pady=(10,0))
entry_pass = tk.Entry(window, show="*")
entry_pass.pack(fill="x", padx=20)

btn_submit = tk.Button(window, text="Gửi")

entry_name.bind("<KeyRelease>", lambda e: kiem_tra())
entry_id.bind("<KeyRelease>", lambda e: kiem_tra())
entry_pass.bind("<KeyRelease>", lambda e: kiem_tra())

window.mainloop()
```

### Bài 7.6.

```python
import tkinter as tk

def dao_chuoi():
    text = entry.get()
    result.set(text[::-1])

window = tk.Tk()
window.title("Reverse String")

tk.Label(window, text="Enter a word:").pack(pady=5)

entry = tk.Entry(window, width=30)
entry.pack()

result = tk.StringVar()
tk.Label(window, textvariable=result).pack(pady=10)

tk.Button(window, text="Validate", command=dao_chuoi).pack()

window.mainloop()
```

```python
#b.
import tkinter as tk

def tim_uoc():
    n = int(entry.get())
    uoc = [str(i) for i in range(1, n + 1) if n % i == 0]
    result.set(" ".join(uoc))

window = tk.Tk()
window.title("Divisors of N")

tk.Label(window, text="Enter the value of N").pack()
entry = tk.Entry(window)
entry.pack()

result = tk.StringVar()
tk.Label(window, text="The divisors of N:").pack()
tk.Label(window, textvariable=result).pack()

tk.Button(window, text="Validate", command=tim_uoc).pack()

window.mainloop()
```

### Bài 7.9.

```python
import tkinter as tk
import math

def tinh():
    m = int(entry_m.get())
    n = int(entry_n.get())
    if choice.get() == "GCD":
        result.set(f"GCD(m, n) = {math.gcd(m, n)}")
    else:
        lcm = abs(m * n) // math.gcd(m, n)
        result.set(f"LCM(m, n) = {lcm}")

window = tk.Tk()
window.title("GCD & LCM")

tk.Label(window, text="Enter value of m").pack()
entry_m = tk.Entry(window)
entry_m.pack()

tk.Label(window, text="Enter value of n").pack()
entry_n = tk.Entry(window)
entry_n.pack()

choice = tk.StringVar(value="GCD")
tk.OptionMenu(window, choice, "GCD", "LCM").pack()

result = tk.StringVar()
tk.Label(window, textvariable=result).pack(pady=5)

tk.Button(window, text="Calculate", command=tinh).pack()

window.mainloop()
```

### Bài 7.11.

```python
import tkinter as tk
from tkinter import messagebox

can = ["Canh","Tân","Nhâm","Quý","Giáp","Ất","Bính","Đinh","Mậu","Kỷ"]
chi = ["Thân","Dậu","Tuất","Hợi","Tý","Sửu","Dần","Mão","Thìn","Tỵ","Ngọ","Mùi"]

def tinh_nam_am():
    year = int(entry.get())
    nam_am = can[year % 10] + " " + chi[year % 12]
    messagebox.showinfo("Kết quả", f"Năm âm lịch: {nam_am}")

window = tk.Tk()
window.title("Đổi năm âm lịch")

tk.Label(window, text="Nhập năm dương lịch").pack()
entry = tk.Entry(window)
entry.pack()

tk.Button(window, text="Chuyển đổi", command=tinh_nam_am).pack(pady=10)

window.mainloop()
```

### Bài 7.12.

```python
import tkinter as tk

def tinh_bmi():
    weight = float(entry_weight.get())
    height = float(entry_height.get())
    bmi = weight / (height ** 2)
    bmi_value.set(f"BMI: {bmi:.2f}")

    if bmi < 18.5:
        result.set("Kết luận: Gầy")
    elif bmi < 25:
        result.set("Kết luận: Bình thường")
    elif bmi < 30:
        result.set("Kết luận: Thừa cân")
    else:
        result.set("Kết luận: Béo phì")

window = tk.Tk()
window.title("BMI Calculator")

tk.Label(window, text="Cân nặng (kg):").pack()
entry_weight = tk.Entry(window)
entry_weight.pack()

tk.Label(window, text="Chiều cao (m):").pack()
entry_height = tk.Entry(window)
entry_height.pack()

tk.Button(window, text="Tính BMI", command=tinh_bmi).pack(pady=5)

bmi_value = tk.StringVar()
tk.Label(window, textvariable=bmi_value).pack()

result = tk.StringVar()
tk.Label(window, textvariable=result).pack()

window.mainloop()
```

### Bài 7.13.

```python
import tkinter as tk
from tkinter import messagebox

def show_info():
    messagebox.showinfo(
        "Lập trình hướng đối tượng",
        "Viết phương trình giải phương trình bậc 2"
    )

window = tk.Tk()
window.title("Hệ thống bài tập nâng cao Python")
window.geometry("500x300")

menu_bar = tk.Menu(window)

for chap in range(1, 5):
    chap_menu = tk.Menu(menu_bar, tearoff=0)
    for i in range(1, 10):
        if chap == 1 and i == 1:
            chap_menu.add_command(label=f"{chap}.{i}", command=show_info)
        else:
            chap_menu.add_command(label=f"{chap}.{i}")
    menu_bar.add_cascade(label=f"Chương {chap}", menu=chap_menu)

menu_bar.add_command(label="Exit", command=window.quit)

window.config(menu=menu_bar)
window.mainloop()
```
