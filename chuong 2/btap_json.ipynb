4. Khái niệm JSON là gì?
- JSON (JavaScript Object Notation) là định dạng trao đổi dữ liệu nhẹ (lightweight data format).
- Dễ đọc, dễ ghi cho con người và dễ phân tích cho máy tính.
- Thường được dùng để truyền dữ liệu giữa client – server trong web.
- Ví dụ:
{
  "ho_ten": "Nguyen Van A",
  "tuoi": 20
}

5. Hạn chế của việc triển khai JSON trong Python
- Không lưu được kiểu dữ liệu phức tạp như set, tuple, object trực tiếp (phải chuyển đổi sang dạng cơ bản).
- Dữ liệu JSON chỉ hỗ trợ kiểu cơ bản: chuỗi, số, danh sách, đối tượng, giá trị true/false/null.
- Không hỗ trợ chú thích (comment).
- Khi dữ liệu quá lớn, tốc độ xử lý có thể chậm hơn so với các định dạng nhị phân (binary).

6. API XML là gì?
- API XML (XML Application Programming Interface) là giao diện lập trình cho phép các chương trình đọc, ghi, phân tích, xử lý dữ liệu XML.
- Python có nhiều API hỗ trợ XML như:
    + xml.dom
    + xml.sax
    + xml.etree.ElementTree
=> Giúp ta thao tác với tài liệu XML theo nhiều cách khác nhau (dạng DOM hoặc dạng sự kiện SAX).

7. API JSON là gì?
- API JSON là giao diện lập trình giúp ta chuyển đổi qua lại giữa đối tượng Python và dữ liệu JSON.

8. Để phân tích tập tin XML trong Python có thể sử dụng các thư viện nào?
- Các thư viện thường dùng:
    + xml.etree.ElementTree – dễ dùng nhất, phổ biến nhất 
    + xml.dom.minidom – xử lý XML theo mô hình DOM.
    + xml.sax – xử lý XML theo mô hình sự kiện (đọc lần lượt từng phần tử).
    + lxml – thư viện mạnh mẽ hơn, hỗ trợ cả XPath và XSLT.
    + BeautifulSoup – thường dùng để trích xuất dữ liệu XML/HTML từ web.

```python
import json
# Chuỗi JSON
chuoi_json = '{"ten": "An", "tuoi": 20, "que": "Ha Noi"}'
# Chuyển chuỗi JSON thành đối tượng Python (từ điển)
doi_tuong = json.loads(chuoi_json)
print("Kết quả sau khi chuyển đổi:")
print(doi_tuong)
print("Tên:", doi_tuong["ten"])
```

Bài 2.8: Chuyển đổi đối tượng Python → chuỗi JSON

```python
import json
# Tạo đối tượng Python
doi_tuong = {"ten": "Bình", "tuoi": 22, "nghe": "Sinh vien"}
# Chuyển thành chuỗi JSON
chuoi_json = json.dumps(doi_tuong)
print("Chuỗi JSON kết quả:")
print(chuoi_json)
```

Bài 2.9: Chuyển đổi từ điển (sắp xếp theo khóa) → JSON có thụt lề

```python
import json
# Tạo từ điển Python
tu_dien = {"c": 3, "a": 1, "b": 2}
# Chuyển đổi sang JSON (sắp xếp theo khóa, thụt lề 4)
chuoi_json = json.dumps(tu_dien, sort_keys=True, indent=4)
print("Kết quả JSON:")
print(chuoi_json)
```

Bài 2.10: Thống kê nhân viên theo đơn vị (từ JSON)

```python
import json
# Đọc dữ liệu JSON
with open("bai2_10.json", "r", encoding="utf-8") as tep:
    du_lieu = json.load(tep)
tong_nv = sum(dv["so_nv"] for dv in du_lieu["don_vi"])
print("Tên công ty:", du_lieu["ten_cong_ty"])
print("Địa chỉ:", du_lieu["dia_chi"])
print("\n--- Thống kê nhân viên theo đơn vị ---")
for i, dv in enumerate(du_lieu["don_vi"], 1):
    ti_le = dv["so_nv"] / tong_nv * 100
    print(f"{i}. Tên đơn vị: {dv['ten_don_vi']}")
    print(f"   Số nhân viên: {dv['so_nv']}")
    print(f"   Tỷ lệ: {ti_le:.2f}%")
```

Bài 2.12 – Đọc và hiển thị các bài post từ API

```python
import requests  # Thư viện chuyên để gọi dữ liệu từ Internet
# Địa chỉ API chứa danh sách các bài post
api_url = "https://jsonplaceholder.typicode.com/posts"
# Gửi yêu cầu (request) đến API và lấy dữ liệu
phan_hoi = requests.get(api_url)
# Chuyển dữ liệu JSON thành danh sách đối tượng Python
danh_sach_post = phan_hoi.json()
# In ra tổng số bài post
print("Tổng số bài post:", len(danh_sach_post))
print("-" * 60)
# In ra từng bài post
for post in danh_sach_post:
    print(f"userId: {post['userId']}")
    print(f"id: {post['id']}")
    print(f"title: {post['title']}")
    print(f"body: {post['body']}")
    print("-" * 60)
```

Bài 2.13 – Đọc và hiển thị “sách nổi bật” (bài post nổi bật)

```python
import requests
# Địa chỉ API của bài post nổi bật
api_url = "https://jsonplaceholder.typicode.com/comments?postId=1"
# Gửi yêu cầu đến API và lấy dữ liệu
phan_hoi = requests.get(api_url)
danh_sach_noi_bat = phan_hoi.json()
print(" Danh sách các bài post nổi bật (giới hạn 3 bài):")
print("-" * 60)
# Duyệt 3 phần tử đầu tiên
for bai in danh_sach_noi_bat[:3]:
    print(f"ID: {bai['id']}")
    print(f"Tên người bình luận: {bai['name']}")
    print(f"Email: {bai['email']}")
    print(f"Nội dung: {bai['body']}")
    print("-" * 60)
```
