ví dụ: 3.4.1

```python
import numpy as np 
import time 
# Tạo một mảng NumPy với 100.000.000 phần tử 
arr = np.arange(100000000, dtype=np.int64) 
# Tính tổng các phần tử của mảng bằng vòng lặp 
start_time = time.time() 
total = np.int64(0) 
for num in arr: 
    total += num 
end_time = time.time() 
elapsed_time_loop = end_time - start_time 
print("Tổng các phần tử mảng arr khi sử dụng vòng lặp:", total) 
print("Thời gian (giây) khi sử dụng vòng lặp:", elapsed_time_loop) 
# Tính tổng các phần tử của mảng bằng np.sum() 
start_time = time.time() 
total = np.sum(arr) 
end_time = time.time() 
elapsed_time_npsum = end_time - start_time 
print("Tổng các phần tử mảng arr khi sử dụng hàm np.sum():", total) 
print("Thời gian (giây) khi sử dụng np.sum():", elapsed_time_npsum) 
num_times_faster = elapsed_time_loop / elapsed_time_npsum 
print("Sử dụng vòng lặp chậm hơn: ~",int(num_times_faster),"lần") 
```

```python
import numpy as np 
 
arr1 = np.array([1, 2, 3]) 
arr2 = np.array([4, 5, 6]) 
 
#Phép Cộng:Sử dụng hàm np.add() hoặc toán tử '+'. 
Tong = np.add(arr1,arr2) 
#hoặc tong = arr1+arr2 
 
#Phép Trừ: Sử dụng hàm np.subtract() hoặc toán tử '-'. 
Hieu = np.subtract(arr1,arr2) 
#hoặc Tru = arr1-arr2 
 
#Phép Nhân: Sử dụng hàm np.multiply() hoặc toán tử '*' 
Tich = np.multiply(arr1,arr2) 
#Hoặc arr1*arr2 
 
#Phép Chia: Sử dụng hàm np.divide() hoặc toán tử '/' 
Thuong = np.divide(arr1,arr2) 
# Về sin cos tan, mp.sin(), np.cos(), np.tan, np.log(), np.log()^2,np.sprt()
 
print("Tổng của arr1 + arr2 là:",Tong) 
print("Hiệu của arr1 - arr2 là:",Hieu) 
print("Tích của arr1 * arr2 là:",Tich) 
print("Thương của arr1/arr2 là:",Thuong)
```
